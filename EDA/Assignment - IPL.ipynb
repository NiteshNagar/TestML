Q14. Bar chart for top 20 players with best strike rate in a season.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ipl=pd.read_csv("IPL_Dataset.csv")

Req_dataSet = ipl[['Player','SR']]
sorted_data = Req_dataSet.sort_values(by='SR', ascending=False).head(20)

# Create a bar chart
plt.figure(figsize=(12, 6))   
plt.bar(sorted_data['Player'], sorted_data['SR'])
plt.xlabel('Players')
plt.ylabel('Strike Rates')
plt.title('Top 20 Players with Best Strike Rates')
plt.xticks(rotation=80)
plt.tight_layout()

# Display the bar chart
plt.show()
```

Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ipl2 = pd.read_csv("IPL_Dataset.csv")
ipl = ipl2.head(10)
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ipl['BF'], ipl['Runs'], color='blue', alpha=0.7)
plt.xlabel('Balls Faced')
plt.ylabel('Runs')
plt.title('Runs Scored vs Balls Faced')
plt.grid(True)

for i, player in enumerate(ipl['Player']):
    plt.annotate(player, (ipl['BF'][i], ipl['Runs'][i]), textcoords="offset points", xytext=(10,10), ha='center')

plt.tight_layout()

# Display the scatter plot
plt.show()
```
