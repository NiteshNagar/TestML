Q1. Check how many duplicate rows are present.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loan=pd.read_excel("loan_sanction.xlsx")

loan = loan.drop (['Sr'],axis=1)
loan = loan[['Loan_ID','ApplicantIncome']]
duplicate_rows = loan[loan.duplicated(keep=False)]

if not duplicate_rows.empty:
    print("Duplicate rows:", len(duplicate_rows))
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

q2. If any Duplicates Present, drop those duplicates

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loan=pd.read_excel("loan_sanction.xlsx")

loan = loan.drop (['Sr'],axis=1)
loan = loan[['Loan_ID','ApplicantIncome']]

drop_duplicates = loan.drop_duplicates(keep='first')
drop_duplicates.tail(5)
```
